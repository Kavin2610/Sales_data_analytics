Load the Dataset
    

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##Load the dataset
df = pd.read_csv(r"C:\Users\Kavin\Desktop\Data_Analytics\sales_data.csv.csv")
df.head()
```

```python
#dropping unwanted columns
df.drop(columns = "Invoice ID", inplace = True)
```

```python
##Question 1 : list the sales volume from top to bottom according to the city
city_sales = df.groupby('City')['Total'].sum()
```

```python
print(city_sales)
```

```python
plt.figure(figsize=(10, 6))
city_sales.plot(kind='bar')

# Adding title and labels
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')

# Displaying the plot
plt.show()
```

```python
# Plot a graph with the highest selling branch according to the quantity
higest_selling_branch = df.groupby('Branch')['Quantity'].sum()
print(higest_selling_branch)
```

```python
plt.figure(figsize=(10, 6))
higest_selling_branch.plot(kind="bar")
plt.title = ('Best selling branch')
plt.xlabel = ('Branch')
plt.ylabel = ('Total Sales')

plt.show()
```

```python
## find out who spends more MALE/FEMALE
more_spenders = df.groupby('Gender')['Total'].sum()
```

```python
print(more_spenders)
```

```python
more_spenders.plot(kind="bar")
plt.title = "More Spenders"
```

```python
##check who purchase more Member/Normal
Customer_type_purchase = df.groupby('Customer type')['Total'].sum()
```

```python
print(Customer_type_purchase)
```

```python
plt.figure(figsize=(10, 6))
Customer_type_purchase.plot(kind='bar')


# Displaying the plot
plt.show()
```

```python
##find out which payment mode has received more amount of money

highest_payment_mode = df.groupby('Payment')['Total'].sum()
print(highest_payment_mode)
```

```python
highest_payment_mode.plot(kind="bar")
```

```python
best_rating = df.groupby('Branch')['Rating'].mean()
print(best_rating)
```

```python
best_rating.plot(kind = "bar")
plt.show()
```

```python
highest_earning_branch = df.groupby('Branch')['gross income'].sum()
print(highest_earning_branch)
```

```python
highest_earning_branch.plot(kind="bar")
```
